```python
import pandas as pd
import numpy as np
import sklearn
from sklearn import datasets
from sklearn.datasets import load_boston
from sklearn.preprocessing import scale
from sklearn.model_selection import KFold
from sklearn.neighbors import KNeighborsRegressor
from sklearn.model_selection import cross_val_score
```

```python
X, y = load_boston(return_X_y=True)
X_scaled = scale(X)
```

```python
kf = KFold(n_splits=5, shuffle=True, random_state=42)
p_range = np.linspace(1, 10, 200)
p_scores = []
for p in p_range:
    knr = KNeighborsRegressor(n_neighbors=5, weights='distance', p=p)
    quality_scores = cross_val_score(knr, X, y, cv=kf, scoring='neg_mean_squared_error')
    p_scores.append(quality_scores.mean())
print(p_scores)
```

```python
print('Length of list', len(p_scores))
print('Max of list', max(p_scores))
print('Optimal p', p_scores.index(max(p_scores))+1)
```
